```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values_pos = np.array([])
x_values_neg = np.array([])
alpha_values = np.arange(-0.999,1,0.001)
```

```python
alpha_values
```

```python
x_values_pos = np.sqrt(alpha_values)
x_values_neg = np.negative(np.sqrt(alpha_values))
```

```python
fig, ax = plt.subplots(figsize=(6,5))

ax.plot(alpha_values, x_values_pos, color='blue', label="Unstable steady states")
ax.plot(alpha_values, x_values_neg, color='blue')
ax.scatter(0, 0, color='red', label="Stable steady states")

plt.xlim([-1,1])
ax.set_xlabel("α")
ax.set_ylabel("x")

plt.legend()
plt.savefig("task2_diagram1.png")
plt.show()
```
